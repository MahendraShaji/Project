```python
from bs4 import BeautifulSoup
import urllib
import requests
import pandas as pd
```

```python
# credits- Thanks Roshan for the idea and layout .
# I have implemented the scrapping through beautifulsoup
job_desc_i_want = ['Data-Scientist', 'Data-Analyst', 'Business-Intelligence', 'Research-Scientist','Data-Architect']
where_is_the_job = ['-jobs/in-All-Sydney-NSW','-jobs/in-All-Melbourne-VIC','-jobs/in-All-Brisbane-QLD']
search_urls = []
for i in range(0,len(where_is_the_job)):
    for index in range(0, len(job_desc_i_want)):
        job_dict = {}
        job_dict['type'] = job_desc_i_want[index]
        job_dict['page'] = 1
        job_dict['url'] = 'https://www.seek.com.au/' + job_desc_i_want[index] + where_is_the_job[i]
        search_urls.append(job_dict)
```

```python
pagelinks_df = pd.DataFrame(columns = ('Job_Desc', 'Page_Num', 'url_link'))
```

```python
# Define the start request function
def get_all_page_links_per_job(url, job_desc, page_num=1):
    # Get the html response from the first page url
    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser', from_encoding="utf-8")
    # get the link to the next page
    next_page_anchor=''
    for a in soup.find_all('a',{'data-automation':'page-next'}, href=True):
        next_page_anchor=a['href']
    
    # check if there is a next page
    if len(next_page_anchor) > 0:
        # Save the current url details before finding the next one
        pagelinks_df.loc[len(pagelinks_df),:] = [job_desc, page_num, url]
        # Start searching for next page link
        # Recursively call to get the links in next anchor page
        page_num += 1
        next_page_anchor = "https://www.seek.com.au" + next_page_anchor
        get_all_page_links_per_job(next_page_anchor, job_desc, page_num)
    else:
        pass
```

```python

def parse_all_first_pages(urls=search_urls):
    for url in urls:
        get_all_page_links_per_job(url['url'], url['type']) 

#Parse all the required job search results and get the links to the job postings
parse_all_first_pages()
```

```python
# Display the dataframe obtained
pagelinks_df.head()
```

```python
# Get all the job posting links from each page
list_of_posts = []
for index in range(0, len(pagelinks_df)):
    # Get the response text from the url
    url = pagelinks_df.iloc[index, 2]
    html = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser', from_encoding="utf-8")
    post=[]
    for entry in soup.find_all('a',{'class':'_1OFaluu'},href=True):
        post.append(entry['href'])
    list_of_posts.append(post)

pagelinks_df['Job_Posts'] = list_of_posts
```

```python
# Go through each page link and get the job post content:
# Description, Job_Title, Salary, Location, Work_Type, Classification,
# DatePosted, Postedby
# Define the start request function
rows = []
for index in range(0, len(pagelinks_df)):
    # Get the response text from the url
#     print('\nPage', index)
    urls = pagelinks_df.iloc[index, 3]
    keyword = pagelinks_df.iloc[index, 0]
    pagenum = pagelinks_df.iloc[index, 1]
    count = 0
    print('\n No of urls to parse:', len(urls))
    for url in urls:
        count = count + 1
#         print("\nParshing url: ", count, '/', len(urls))
        url = 'https://www.seek.com.au' + url
        try:
            html = urllib.request.urlopen(url).read()
            soup = BeautifulSoup(html, 'html.parser', from_encoding="utf-8")
            if len(soup.find_all('h1',{'class':'jobtitle'}))>0:
                title=soup.find('h1',{'class':'jobtitle'}).renderContents()
            else:
                title=''
            advertiser=soup.find('span',{'class':'state-message'}).renderContents().decode('ISO-8859-1')
            if len(soup.find_all('div',{'itemprop':'baseSalary'}))>0:
                salary=soup.find('div',{'itemprop':'baseSalary'}).renderContents()
            else:
                salary=''
            if len(soup.find_all('div',{'itemprop':'employmentType'}))>0:
                worktype=soup.find('div',{'itemprop':'employmentType'}).renderContents()
            else:
                worktype=''
            if len(soup.find_all('span',{'itemprop':'industry'}))>0:
                classification=soup.find('span',{'itemprop':'industry'}).renderContents()
            else:
                classification=''
            if len(soup.find_all('span',{'itemprop':'addressLocality'}))>0:
                location=soup.find('span',{'itemprop':'addressLocality'}).renderContents()
            else:
                location=''
            if(len(soup.find_all('span',{'itemprop':'addressRegion'}))>0):
                p=soup.find('span',{'itemprop':'addressRegion'})
                for child in p.children:
                    if child.name!="span":
                        region=child
            else:
                region=''

            jd=[]
            if(len(soup.find_all('div',{'class':'templatetext'}))>0):
                p=soup.find('div',{'class':'templatetext'})
                for child in p.children:
                    jd.append(child)
            else:
                jd=''
            # Append the scraped items to the list of rows
            rows.append([keyword, pagenum, title, location, region, salary, worktype, advertiser, classification, jd, url])
        except requests.exceptions.ConnectionError:
            pass
```

```python
columns = ('keyword', 'pagenumber', 'title', 'location','region', 'salary', 'worktype', 'advertiser', 'classification', 'jd', 'url')
job_posts_df = pd.DataFrame(rows, columns = columns)
```
